```python
import numpy as np
import matplotlib.pyplot as plt

N1 = 100
N2 = 100
mu1 = 0.5
mu2 = 1
var1 = 0.5
var2 = 0.6

X1 = np.random.normal(mu1, var1, N1)
X2 = np.random.normal(mu2, var2, N2)

muy1 = 0.5
vary1 = 0.04
y1 = np.random.normal(muy1, vary1, N1)

muy2 = 1.5
vary2 = 0.05
y2 = np.random.normal(muy2, vary2, N2)

plt.title('P(X,Y)')
plt.scatter(X1, y1)
plt.scatter(X2, y2)

# Añadir rejillas al gráfico
plt.grid(True, which='both', axis='both', color='blue', linestyle='--', linewidth=1, alpha=1)

# Personalizar las rejillas para tener 9 en el eje x y 2 en el eje y
plt.xticks(np.linspace(plt.xlim()[0], plt.xlim()[1], 9))
plt.yticks(np.linspace(plt.ylim()[0], plt.ylim()[1], 3))

plt.show()
```

```python
# Valores de X
N_1 = 100
N_2 = 100
mu_1x = 0.5
mu_2x = 2
var_1x = 0.8
var_2x = 0.6

# Valores de Y
mu_1y = 0.5
mu_2y = 1.5
var_1y = 0.15
var_2y = 0.15
```

```python
X_1 = np.random.normal(mu_1x, var_1x, N_1)
X_2 = np.random.normal(mu_2x, var_2x, N_2)
Y_1 = np.random.normal(mu_1y, var_1y, N_1)
Y_2 = np.random.normal(mu_2y, var_2y, N_2)
```

```python
plt.scatter(X_1, Y_1)
plt.scatter(X_2, Y_2)
```

```python
def proba_param (X1, X2, Y1, Y2):
  plt.subplot(2, 2, 1)
  plt.scatter (X1, Y1)
  plt.scatter (X2, Y2)

  plt.subplot(2, 2, 2)
  plt.hist ([np.concatenate((Y1,Y2))], bins = 2, alpha = 0.5)

  plt.subplot(2, 2, 3)
  plt.hist ([np.concatenate((X1, X2))], bins = 20, alpha = 0.5)
  plt.xlim(-2 , 5)

  plt.subplot(2, 2, 4)
  plt.hist (X2, bins = 9, alpha = 0.5)
  plt.xlim(-2, 5)

  plt.show()
```

```python
proba_param(X_1, X_2, Y_1, Y_2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0
sigma = 1
num_samples = 1000

# Generar muestras aleatorias de la distribución gaussiana
data = np.random.normal(mu, sigma, num_samples)

# Crear el histograma
plt.hist(data, bins=30, density=True, alpha=0.6, color='b')

# Calcular la función de densidad de probabilidad teórica
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Graficar la función de densidad de probabilidad
plt.plot(x, pdf, color='r', linewidth=2)

# Mostrar el gráfico
plt.title('Histograma de una distribución gaussiana')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N1 = 50
mu1 = 0.5
var1 = 0.8

# Generar datos para la distribución normal X1
X1 = np.random.normal(mu1, var1, N1)

# Generar histograma para la variable X1
plt.hist(X1, bins=9, density=True, alpha=0.6, color='b')

# Mostrar el gráfico
plt.title('Histograma P(X)')
plt.xlabel('Valor de X1')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
